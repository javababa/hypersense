###### We can drop the Assembly column since it is constant. (Kolkata Port)

```python
# Dropping the `Assembly` column which has no utility in this particular problem statement.

elections.drop("Assembly", axis=1, inplace=True)

# Dropping the `Sanatani_D` column which has no utility in this particular problem statement.

elections.drop("Sanatani_D", axis=1, inplace=True)

# Display the Remaining Columns.
elections.info()
```

```python
# Find Total Number of Voters in the constituency.
total_electors = elections['TotalVoters'].sum()
total_male = elections['Male'].sum()
total_female = elections['Female'].sum()
total_trans = elections['Transgender'].sum()
total_sanatani = elections['SanataniVoters'].sum()
total_abrahamic = elections['AbrahamicVoters'].sum()

# Printing the Total Voter Categories
print("Total Voters: ", total_electors)
print("Total Males: ", total_male)
print("Total Females: ", total_female)
print("Total Transgenders: ", total_trans)
print("Total Sanatani: ", total_sanatani)
print("Total Abrahamic: ", total_abrahamic)
```

```python
# Print the Percentages of the brouad category of voters.

# Male Voter Percentage
print("Percentage Male Voters: {} %".format((total_male*100/total_electors).round(2)))

# Female Voter Percentage
print("Percentage Female Voters: {} %".format((total_female*100/total_electors).round(2)))

# Sanatani Voter Percentage
print("Percentage Sanatani Voters: {} %".format((total_sanatani*100/total_electors).round(2)))

# Abrahamic Voter Percentage
print("Percentage Abrahamic Voters: {} %".format((total_abrahamic*100/total_electors).round(2)))
```

```python
# Find the aggregate of 'SanataniVoters' seggregated by Wards.
sanatani_voters_by_wards = elections.groupby("Ward")['SanataniVoters'].agg(["sum", "mean", "std"]).round(0)

# Rename the Columns.

sanatani_voters_by_wards.rename(columns={'sum': "Total Sanatani Voters", "mean": "Average Sanatani Voters Per Booth", "std": "Sanatani Std Dev."}, inplace=True)
```

```python
# Find the aggregate of 'AbrahamicVoters' seggregated by Wards.

abrahamic_voters_by_wards = elections.groupby("Ward")['AbrahamicVoters'].agg(["sum", "mean", "std"]).round(0)

# Rename the Columns.

abrahamic_voters_by_wards.rename(columns={'sum': "Total Abrahamic Voters", "mean": "Average Abrahamic Voters Per Booth", "std": "Abrahamic Std Dev."}, inplace=True)
```

```python
# Join the above DataFrames and Find the Ratio of Sanatani To Abrahamic in each Ward

ward_wise_sanatan_abrahamic_params = sanatani_voters_by_wards.join(abrahamic_voters_by_wards)

# We want to have 3 columns Total Sanatani Voters, Total Abrahamic Voters and Sanatani / Abrahamic Ratio

# So we first choose to drop unnecessary columns.
ward_wise_sanatan_abrahamic_ratios = ward_wise_sanatan_abrahamic_params.drop(['Average Sanatani Voters Per Booth', 'Average Abrahamic Voters Per Booth', 'Sanatani Std Dev.', 'Abrahamic Std Dev.'], axis = 1)
ward_wise_sanatan_abrahamic_ratios['Sanatani / Abrahamic Ratio'] = (ward_wise_sanatan_abrahamic_ratios['Total Sanatani Voters'] / ward_wise_sanatan_abrahamic_ratios['Total Abrahamic Voters']).round(4)

ward_wise_sanatan_abrahamic_ratios['Sanatani Percentage'] =         (ward_wise_sanatan_abrahamic_ratios['Total Sanatani Voters'] * 100 / (ward_wise_sanatan_abrahamic_ratios['Total Sanatani Voters'] + ward_wise_sanatan_abrahamic_ratios['Total Abrahamic Voters'])).round(2)

ward_wise_sanatan_abrahamic_ratios['Abrahamic Percentage'] =         (ward_wise_sanatan_abrahamic_ratios['Total Abrahamic Voters'] * 100 / (ward_wise_sanatan_abrahamic_ratios['Total Sanatani Voters'] + ward_wise_sanatan_abrahamic_ratios['Total Abrahamic Voters'])).round(2)
ward_wise_sanatan_abrahamic_ratios
```

```python
# Wardwise Total Vote Distribution of BJP, TMC and INC_CPIM_ISF

agg_total_votes_all = elections.groupby('Ward')[['BJP', 'TMC', 'INC_CPIM_ISF']].agg("sum")
```

```python
# Wardwise Mean Vote per Booth Distribution of BJP, TMC and INC_CPIM_ISF

agg_mean_votes_all = agg_mean_votes_all = elections.groupby('Ward')[['BJP', 'TMC', 'INC_CPIM_ISF']].agg("mean").round(0)
```

```python
# Wardwise Std. DEV of Distribution of BJP, TMC and INC_CPIM_ISF

agg_std_dev_votes_all = elections.groupby('Ward')[['BJP', 'TMC', 'INC_CPIM_ISF']].agg("std").round(0)
```
